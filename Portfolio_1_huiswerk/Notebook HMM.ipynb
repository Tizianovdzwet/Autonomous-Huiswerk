## Het HMM model

Het trainen van een HMM aan de hand van de gegeven data doen we in deze Notebook na het genereren met een gegeven model. We gebruiken de package `hmmlearn`. 

Het HMM lijkt op een Markov Chain, maar heeft een extra laag van verborgen states. De verborgen states zijn de toestanden die we niet kunnen zien, maar waarvan we wel de overgangen kunnen zien. De observaties zijn de toestanden die we wel kunnen zien. 

In dit geval zijn de verborgen toestanden de weersomstandigheden en de observaties de temperatuur. De data bestaat uit een lijst van activiteiten. Het is waarschijnlijk dat de weersomstandigheden invloed hebben op de activiteiten die mensen ondernemen. 

```python
import numpy as np

# Definieer de HMM-parameters
weer = ['Zon', 'Regen']  # Verborgen toestanden
activiteiten = ['Wandelen', 'Winkelen', 'Schoonmaken']  # Mogelijke observaties
```
