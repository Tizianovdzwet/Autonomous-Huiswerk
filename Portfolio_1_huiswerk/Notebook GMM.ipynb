# Data laden

In deze Notebook wordt een Gaussian Mixture Model (GMM) gebruikt om een dataset van handgeschreven cijfers te analyseren en nieuwe cijfers te genereren. 

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

De dataset van handgeschreven cijfers (`digits`) wordt geladen en een functie `plot_digits` wordt gedefinieerd om de cijfers te visualiseren. De functie maakt een 10x10 raster van subplots en toont de eerste 100 cijfers uit de dataset.

```python
from sklearn.datasets import load_digits
from sklearn.utils import Bunch
digits: Bunch = load_digits() # type: ignore

def plot_digits(data):
    fig, ax = plt.subplots(10, 10, figsize=(8, 8), subplot_kw=dict(xticks=[], yticks=[]))
    fig.subplots_adjust(hspace=0.05, wspace=0.05)
    for i, axi in enumerate(ax.flat):
        im = axi.imshow(data[i].reshape(8, 8), cmap='binary')
        im.set_clim(0, 16)
plot_digits(digits.data)
```
